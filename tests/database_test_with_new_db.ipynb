CHECK DATABASE INTEGRITY AND TABLES

```python
import sqlite3
import pandas as pd
import os
from pathlib import Path
# current_dir = Path(__file__).resolve().parent #not working in jupyter notebook
current_dir = Path(os.getcwd())

parent_dir = current_dir.parent
db_path = parent_dir / "data" / "database.db"

def run_command(query):
    try:
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        cursor.execute(query)
        results = cursor.fetchall()
        conn.close()
        return results
    except sqlite3.Error as e:
        print(f"An error occurred: {e}")
        return None

run_command("SELECT name FROM sqlite_master WHERE type='table';")
```
